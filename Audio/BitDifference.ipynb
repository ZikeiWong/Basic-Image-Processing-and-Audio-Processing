```python
import librosa
import librosa.display
import IPython.display as ipd
import numpy as np

y,sr=librosa.load('voice.wav')

b1=4
N1=2**(b1-1)
y1=np.round(y*N1)/N1

b2=8
N2=2**(b2-1)
y2=np.round(y*N2)/N2

b3=16
N3=2**(b3-1)
y3=np.round(y*N3)/N3

import matplotlib.pyplot as plt
plt.figure()
librosa.display.waveplot(y1)
plt.title('4bits')

plt.figure()
librosa.display.waveplot(y2)
plt.title('8bits')

plt.figure()
librosa.display.waveplot(y3)
plt.title('16bits')

ipd.display(ipd.Audio(y1,rate=sr))
ipd.display(ipd.Audio(y2,rate=sr))
ipd.display(ipd.Audio(y3,rate=sr))
```
