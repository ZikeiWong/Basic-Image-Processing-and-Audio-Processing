```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
x, sr1 = librosa.load('piano.wav')
y, sr2 = librosa.load('voice.wav')
```

```python
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr1, x_axis='time', y_axis='hz')
plt.colorbar()
```

```python
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr1, x_axis='time', y_axis='log')
plt.colorbar()
```

```python
Y = librosa.stft(y)
Ydb = librosa.amplitude_to_db(abs(Y))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Ydb, sr=sr2, x_axis='time', y_axis='hz')
plt.colorbar()
```

```python
plt.figure(figsize=(30, 10))
librosa.display.specshow(Ydb, sr=sr2, x_axis='time', y_axis='log')
plt.colorbar()
```

```python
X = librosa.stft(x,n_fft=4096)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(50, 30))
librosa.display.specshow(Xdb,sr=sr2,cmap='inferno', x_axis='time', y_axis='log')
plt.colorbar()
```
