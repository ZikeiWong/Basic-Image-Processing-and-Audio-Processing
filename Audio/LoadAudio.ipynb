```python
import numpy as np
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt

sr_sin = 44100 #Sample Rate
f = 441
n = np.arange(sr_sin*2)
y_sin = np.sin(2 * np.pi * f / sr_sin * n)

plt.figure(figsize=(100,20))
librosa.display.waveplot(y_sin, sr = sr_sin)
librosa.output.write_wav('441sin.wav', y_sin.astype(np.float32), sr=sr_sin)
ipd.Audio(y_sin, rate = sr_sin)
```
