```python
import numpy as np
import librosa.display
import IPython.display as ipd

sr_sin = 44100 #Sample Rate
f = 441
n = np.arange(sr_sin*2)
y_sin = np.sin(2 * np.pi * f / sr_sin * n)

sin1 = librosa.resample(y_sin,sr_sin,8000)
ipd.Audio(sin1,rate=8000)
```

```python
sin2 = librosa.resample(y_sin,sr_sin,22050)
ipd.Audio(sin2,rate=22050)
```

```python
sin3 = librosa.resample(y_sin,sr_sin,44100)
ipd.Audio(sin3,rate=44100)
```
